```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
```

```python
df = pd.read_excel('stock_data.xlsx', parse_dates = True, index_col = 'Date')
df.tail()
```

```python
df2019 = df.loc['2019-01-01':'2019-12-31']
df2019[['Close','Close5']].plot()
plt.show()
```

```python
df2019.tail()
```

```python
ax1 = plt.subplot2grid((6,1), (0,0), rowspan=5, colspan=1)
ax2 = plt.subplot2grid((6,1), (5,0), rowspan=5, colspan=1, sharex=ax1)

ax1.plot(df2019.index, df2019['Adj Close'])
ax1.plot(df2019.index, df2019['20ma'])
ax2.bar(df2019.index, df2019['Volume'])

plt.show()
```
